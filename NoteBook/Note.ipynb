```python
null_counts = df.isnull().sum()
null_counts = null_counts[null_counts > 0]
null_df = null_counts.reset_index()
null_df.columns = ['Column', 'Null_Count']
null_df = null_df.sort_values(by='Null_Count', ascending=False)
print(null_df)
```

```python
null_percent = (df.isnull().mean() * 100).reset_index()
null_percent.columns = ['Column', 'Null_Percent']
null_percent = null_percent[null_percent['Null_Percent'] > 0]
null_percent = null_percent.sort_values(by='Null_Percent', ascending=False)
print(null_percent)
```
