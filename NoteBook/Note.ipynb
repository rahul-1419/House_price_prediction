```python
import pandas as pd
target = "SalePrice"

# 1. Find correlation matrix
corr_matrix = df.corr(numeric_only=True)

# 2. Find highly correlated features (threshold > 0.8)
threshold = 0.8
high_corr = set()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname1 = corr_matrix.columns[i]
            colname2 = corr_matrix.columns[j]
            high_corr.add((colname1, colname2, corr_matrix.iloc[i, j]))

# 3. Find weak correlation with target (< 0.1)
weak_corr = corr_matrix[target][abs(corr_matrix[target]) < 0.1].index.tolist()

# ---- Results ----
print(" Highly correlated column pairs (>|0.8|):")
for col1, col2, corr_val in high_corr:
    print(f"{col1}  {col2} | Corr = {corr_val:.2f}")

print("\n Weakly correlated columns with target (|corr| < 0.1):")
print(weak_corr)
```

```python
# For categorical vs numerical
from sklearn.feature_selection import f_regression

X = pd.get_dummies(df[categorical_features], drop_first=True)
f_values, p_values = f_regression(X, df['SalePrice'])

anova_results = pd.DataFrame({
    'Feature': X.columns,
    'F_Value': f_values,
    'p_value': p_values
}).sort_values(by='F_Value', ascending=False)

anova_results
```

```python
def correlation_ratio(categories, values):
    # Convert categories to pandas categorical type
    categories = pd.Categorical(categories)
    cat_levels = categories.categories
    
    # Overall mean of target variable
    y_mean = np.mean(values)
    
    # Total variance (denominator)
    ss_total = np.sum((values - y_mean) ** 2)
    
    # Between-group variance (numerator)
    ss_between = 0
    for cat in cat_levels:
        y_cat = values[categories == cat]
        ss_between += len(y_cat) * (np.mean(y_cat) - y_mean) ** 2
    
    # Correlation ratio ()
    eta_squared = ss_between / ss_total if ss_total != 0 else 0
    return eta_squared
```

```python
# Store results
eta_results = {}
for col in categorical_features:
    try:
        eta = correlation_ratio(df[col], df['SalePrice'])
        eta_results[col] = eta
    except Exception as e:
        eta_results[col] = np.nan

# Sort by strongest effect
eta_df = pd.DataFrame(eta_results.items(), columns=['Feature', 'Eta_Squared'])
eta_df = eta_df.sort_values(by='Eta_Squared', ascending=False)
print(eta_df)
```

```python
import matplotlib.pyplot as plt

top_features = eta_df.head(10)
plt.figure(figsize=(8, 5))
plt.barh(top_features['Feature'], top_features['Eta_Squared'], color='skyblue')
plt.xlabel('Correlation Ratio ()')
plt.title('Top 10 Categorical Features Affecting SalePrice')
plt.gca().invert_yaxis()
plt.show()
```
